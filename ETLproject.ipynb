```python
import newspaper
import random
textContent =[]
cnn_paper = newspaper.build('http://cnn.com', memoize_articles=False)
cnn_article = cnn_paper.articles[0]
cnn_article.download()
cnn_article.parse()
cnn_article.nlp()
for article in cnn_paper.articles:
    print(article.url)
    textContent.append(article.url)
textContent=random.choices(textContent,k=5)
```

```python
from textblob import TextBlob
from newspaper import Article
import pandas as pd
sents = []
article_titles = []
paragraphs = []
for item in textContent:
    try:
        url = item
        article = Article(url)
        article.download()
        # html = article.html
        article.parse()
        title = article.title
        blob = TextBlob(article.text)
        sents.append(blob.polarity)
        article_titles.append(title)
        paragraphs.append(article.summary)
    except:
        pass
series = pd.Series(sents)
series.describe()
```

```python
df = pd.DataFrame({'title':article_titles, 'paragraph':paragraphs,'score':sents})
```
